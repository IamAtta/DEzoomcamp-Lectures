```python
file_path = 'yellow_tripdata_2021-01.csv.gz'
```

```python
import pandas as pd

# Define the data type mapping
date_columns = ['tpep_pickup_datetime', 'tpep_dropoff_datetime']
dtype_mapping = {'store_and_fwd_flag': 'str'}

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path, compression='gzip', dtype=dtype_mapping, parse_dates = date_columns)
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
engine.connect()
```

```python
print(pd.io.sql.get_schema(df, name = 'yellow_taxi_data', con=engine))
```

```python
df_iter = pd.read_csv(file_path, compression='gzip', dtype=dtype_mapping, parse_dates = date_columns, iterator= True, chunksize = 100000)
```

```python
df = next(df_iter)
```

```python
len(df)
```

```python
df
```

```python
df.head(n=0).to_sql(name = 'yellow_taxi_data', con = engine, if_exists='replace')
```

```python
%time df.to_sql(name = 'yellow_taxi_data', con = engine, if_exists='append')
```

```python
from time import time
```

```python
while True :

    t_start = time()
    
    df = next(df_iter)
    
    df.to_sql(name = 'yellow_taxi_data', con = engine, if_exists='append')
    
    t_end = time()

    print('inserted another chunk..., took %.3f second' % (t_end - t_start))

```
